```python
## Question 1 ##
```

```python
driver=webdriver.Edge()
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Scientist')

search=driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()

```

```python
salary=driver.find_element(By.XPATH,'//p/span[contains(text(),"3-6")]')
salary.click()

```

```python
location=driver.find_element(By.XPATH,'//p/span[contains(text(),"Bengaluru")]')
location.click()
```

```python
job_titles=[]
title_tag=driver.find_elements(By.XPATH,'//div/a[contains(text(),"Data")]')
for i in title_tag[0:10]:
    tags=i.text
    job_titles.append(tags)

job_location=[]
job_loc=driver.find_elements(By.XPATH,'//span[@class="loc-wrap ver-line"]')
for i in job_loc[0:10]:
    loc=i.text
    job_location.append(loc)

company_name=[]
comp_name=driver.find_elements(By.XPATH,'//div[@class=" row2"]')
for i in comp_name[0:10]:
    comp=i.text
    company_name.append(comp)

exp_required=[]
experiance_req=driver.find_elements(By.XPATH,'//span[@class="exp-wrap"]')
for i in experiance_req[0:10]:
    exp=i.text
    exp_required.append(exp)
```

```python
print(len(job_titles),len(job_location),len(company_name),len(exp_required))
```

```python
df=pd.DataFrame({'Titles':job_titles,'location':job_location,'Company':company_name,'Experiance':exp_required})
df

```
